```python
import numpy as np
import pandas as pd
import plotly.express as px
from scipy.ndimage.filters import gaussian_filter1d
```

```python
states = [
    'CA', # California
    'IL', # Illinois
    'IN', # Indiana
    'IA', # Iowa
    'KS', # Kansas
    'MI', # Michigan
    'MN', # Minnesota
    'MO', # Missouri
    'NE', # Nebraska
    'ND', # North Dakota
    'OH', # Ohio
    'SD', # South Dakota
    'WI'  # Wisconsin
]

months = [
    'jan',
    'feb',
    'mar',
    'apr',
    'may',
    'jun',
    'jul',
    'aug',
    'sep',
    'oct',
    'nov',
    'dic'
]

# Set this as the state index to process each state
currState = 0

df = pd.read_csv('../data/wx/wxsummary.csv')
df.shape[0]
```

```python
# Delete unwated states and keeping
# only the list above

length = df.shape[0]

for i in range(0, length):
    if df['state'][i] not in states:
        df = df.drop([i])
```

```python
# Get state by state and remove NaN values

df = df.loc[df['state'] == states[currState]]

df = df.dropna(inplace = False)
df

```

```python
# Calculate the mean of every month

newDF = df.drop(['type', 'state', 'year'], axis=1)
df['avg'] = newDF.mean(axis=1)

df
```

```python
# Process graph and save image under images/weather

fig = px.scatter(
    df, 
    x="year", 
    y="avg",
    
    color='type', 
    title='Weather data for '+ states[currState] +' 1990 - 2020'

)

fig.write_image('../images/weather/'+ states[currState] +'.png')
```
